####  Solution of Part1

```python
def find_pairs_with_products(arr, n, target_sum):
    # Store counts of all elements in map m
    m = {}
    pairs_with_products = {}

    # Traverse through all elements
    for i in range(n):
        # Search if a pair can be formed with arr[i].
        rem = target_sum - arr[i]
        if rem in m:
            # Calculate the product of the pair
            product = rem * arr[i]
            pairs_with_products[(rem, arr[i])] = product
        m[arr[i]] = m.get(arr[i], 0) + 1

    return pairs_with_products

# Example usage
target_sum = 2020
n = len(input_data_new)
pairs_with_products = find_pairs_with_products(input_data_new, n, target_sum)
print(pairs_with_products)  # Output should be {(2, 5): 10, (3, 4): 12
```

#### Solution of Part2 

```python
def find3Numbers(arr, target_sum):
    triplets_with_products = {}

    # Fix the first element as arr[i]
    for i in range(len(arr) - 2):
        # Create a set to store potential second elements that complement the desired sum
        s = set()
        # Calculate the current sum needed to reach the target sum
        curr_sum = target_sum - arr[i]

        # Iterate through the subarray arr[i+1:]
        for j in range(i + 1, len(arr)):
            # Calculate the required value for the second element
            required_value = curr_sum - arr[j]

            # Check if the required value is present in the set
            if required_value in s:
                # Triplet is found; store the triplet and its product in the dictionary
                triplet = (arr[i], arr[j], required_value)
                product = arr[i] * arr[j] * required_value
                triplets_with_products[triplet] = product

            # Add the current element to the set for future complement checks
            s.add(arr[j])

    return triplets_with_products

# Driver program to test the above function
if __name__ == "__main__":
    target_sum = 2020

    triplets_with_products = find3Numbers(input_data_new, target_sum)
    if not triplets_with_products:
        print("No triplet found.")
    else:
        print("Triplets with their products:", triplets_with_products)
```
